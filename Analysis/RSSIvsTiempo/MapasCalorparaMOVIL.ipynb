## Elige posicion X (1 ... 15)

```python
for beacon in range(5):
    for pos in range(15):
        #Leer CSV  Tx 0x04 toma 1
        t04_t1 = pd.read_csv('./Movil/Tx_0x04/Tx04_3.csv', names=[i for i in range(1,3000)],skiprows=2)
        data = t04_t1.transpose() # cada columna es una posicion del 1 al 15
        data2 = data[pos][3:] # toma la posicion 1, desde la 3ra fila , las 2 primeras es basura
        data3 = data2.as_matrix() # lo convierte a matriz
        data3= [v for v in data3 if not math.isnan(v)] # elimina valores NaN
        data4 = []
        data4 = [ [j for j in data3[i::5] ] for i in range(5)] # filtra por cada beacon (cada 5)
        print(np.mean(data4[beacon])) # RSSI del beacon 07 (1) para posicion 1
    print('\n')
```
